```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.mobafire.com/league-of-legends/items'
browser.visit(url)
```

```python
# create the soup object
html = browser.html
lol_soup = BeautifulSoup(html, 'html.parser')

# Goal - gain a list with all divs that contain hrefs to the information and the photos

#parent div
image_div = lol_soup.find('div', class_='self-clear item-list item-list--list')
```

```python
image_div
```

```python
# parent list for all links
all_links = lol_soup.find_all('a')

counter = 0
for link in all_links:
    link = link.get('href')
    
    print(f'{counter} - {link}')
    counter +=1
```
