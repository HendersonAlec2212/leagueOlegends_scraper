# Begin Image Scraping and Saving

```python
html = browser.html
soup_obj = BeautifulSoup(html, 'html.parser')

all_img_tag = soup_obj.find_all('img')
all_img_tag
```

### clean the image link list

```python
import pymongo
# set up pymongo connection

# https://docs.mongodb.com/manual/reference/default-mongodb-port/

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# declare database
db = client.lol_information

# declare the collection
sr_items = db.sr_items
img_collection = db.img_links
test_items = db.test_items
champions = db.champions
```

```python
# isolate the link in the soup list
# break down each link using positive and negative slicing
# if contains 'item' and 'size in px' append, else pass

link_list = []
base_url = 'https://www.mobafire.com'
for i,link in enumerate(all_img_tag):
    link = all_img_tag[i]['src']
    link_slice = link[8:12]
    link_slice_size = link[-7:-4]

    if (link_slice=='item') and (link_slice_size=='64x'):
        full_link = f'{base_url}{link}'
        link_list.append(full_link)
    else:
        continue
```

```python
# for each link in list
# make dictionary for mongo insertion
# trim string to isolate and create variables

# request image, open and save to scraped images folder

for i,image in enumerate(link_list):
    image_dict = {}
    image_name = link_list[i][37:-8]
    image_link = link_list[i]
    r = requests.get(image_link)
    image = Image.open(io.BytesIO(r.content))
    image_dict[f'{image_name}'] = image_link
    img_collection.insert_one(image_dict)
    image.save(f'scraped_images/{image_name}.png')
```
